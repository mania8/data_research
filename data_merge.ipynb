```python
matching = []
for country_1 in df1["Geopolitical entity (reporting)"].unique():
    for country_2 in df2["Reference area"].unique():
        if country_1==country_2:
            matching.append(country_1)

matching
```

```python
# 1. Tworzymy nową ramkę danych z wierszami, które chcemy zachować
# Używamy .str.strip(), aby usunąć wszelkie wiodące/kończące spacje
# przed wykonaniem DOKŁADNEGO porównania (==).
df2_filtered_by_price = df2_filtered_by_country_and_date[df2_filtered_by_country_and_date['Price base'].str.strip() == 'Chain linked volume (rebased)']

# Powinno wyświetlić tylko ['Chain linked volume']
df2_filtered_by_price['Price base'].unique()

#wyświetlenie
df2_filtered_by_price_sorted = df2_filtered_by_price.sort_values(['Reference area','TIME_PERIOD'])
df2_filtered_by_price_sorted.head(12)

```

```python
import pandas as pd
# Załóżmy, że Twoje DataFrame'y to:
# df1 = df1_filtered_by_country
# df2 = df2_filtered_by_price_sorted

# Wykonanie fuzji (łączenia) na dwóch kluczach: kraju i roku
df_fused = pd.merge(
    df1_filtered_by_country,  # Pierwsza tabela (Eurostat - Czas Pracy)
    df2_filtered_by_price_sorted,  # Druga tabela (OECD - Produktywność)
    
    # Określenie, które kolumny są wspólne (łączenie po tych samych wartościach)
    left_on=['Geopolitical entity (reporting)', 'TIME_PERIOD'],
    right_on=['Reference area', 'TIME_PERIOD'],
    
    # Użycie metody inner join (połącz tylko tam, gdzie są dane w OBU tabelach)
    how='inner' 
)

```

```python
# 1. Zmiana nazw kolumn wartości i kluczy, by były czytelne
df_final = df_fused.rename(columns={
    # Kolumny kluczy
    'Geopolitical entity (reporting)': 'Country',
    'TIME_PERIOD': 'Year',
    
    # Kolumny wartości (OBS_VALUE z sufiksami)
    'OBS_VALUE_x': 'Annual_Hours_Worked',       # Czas Pracy (z df1)
    'OBS_VALUE_y': 'Productivity_PPP_Value',    # Produktywność (z df2)
})

# 2. Wybór tylko 4 potrzebnych kolumn do finalnego Outputu
df_final = df_final[['Country', 'Year', 'Annual_Hours_Worked', 'Productivity_PPP_Value']]

# 3. Sprawdzenie, jak wygląda końcowy DataFrame
print("\n--- Połączony i Końcowy DataFrame ---")
print(df_final.head(20))

# 4. Zapisanie gotowego, ujednoliconego pliku (Output 2: CSV)
df_final.to_csv('final_data_for_analysis.csv', index=False)
```
