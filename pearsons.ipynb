```python
import pandas as pd
from scipy.stats import pearsonr

# Wczytanie końcowego pliku CSV
df_final = pd.read_csv('data/final_data_for_analysis.csv')

# Przykładowe sprawdzenie nazw kolumn po fuzji:
# print(df_final.columns) 

# Upewnij się, że nazwy kolumn do korelacji są poprawne:
hours_column = 'Annual_Hours_Worked'      # Twoja zmienna z Eurostatu
productivity_column = 'Productivity_PPP_Value' # Twoja zmienna z OECD

# Obliczenie korelacji Pearsona
correlation_r, p_value = pearsonr(df_final[hours_column], df_final[productivity_column])

print(f"Współczynnik korelacji Pearsona (r): {correlation_r:.4f}")
print(f"Istotność statystyczna (p): {p_value:.4f}")
```

```python
import matplotlib.pyplot as plt

# Ustawienie rozmiaru wykresu
plt.figure(figsize=(10, 6)) 

# Tworzenie wykresu punktowego
plt.scatter(
    df_final[hours_column], 
    df_final[productivity_column],
    alpha=0.7 
)

# Dodanie etykiet i tytułu
plt.title(f'Korelacja Czas Pracy vs. Produktywność (r = {correlation_r:.2f})')
plt.xlabel('Roczna Liczba Przepracowanych Godzin')
plt.ylabel('Produktywność na Godzinę (PPP)')
plt.grid(True, linestyle='--', alpha=0.6)

# Zapisanie wykresu jako plik PNG (Twój Output 1)
plt.savefig('scatter_plot.png')
plt.show() # Wyświetlenie wykresu w Notebooku
```
