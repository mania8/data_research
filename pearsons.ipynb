```python
import pandas as pd
from scipy.stats import pearsonr

df_final = pd.read_csv('data/combined/final_data_for_analysis.csv')

hours_column = 'Annual_Hours_Worked'      
productivity_column = 'Productivity_PPP_Value' 

correlation_r, p_value = pearsonr(df_final[hours_column], df_final[productivity_column])

print(f"Pearson correlation coefficient (r): {correlation_r:.4f}")
print(f"statistical significance (p): {p_value:.4f}")
```

```python
import matplotlib.pyplot as plt

# size of chart
plt.figure(figsize=(10, 6)) 

# creating scatter plot
plt.scatter(
    df_final[hours_column], 
    df_final[productivity_column],
    alpha=0.7 
)

# title
plt.title(f'Working Time vs. Productivity Correlation (r = {correlation_r:.2f})')
plt.xlabel('Annual Number of Hours Worked')
plt.ylabel('Productivity per Hour (PPP)')
plt.grid(True, linestyle='--', alpha=0.6)

# saving
plt.savefig('scatter_plot.png')
plt.show() # Displaying a chart in Notebook
```
